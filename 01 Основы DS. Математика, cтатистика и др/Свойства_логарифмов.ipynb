# Конспект: Свойства логарифмов

Логарифм — это операция, обратная возведению в степень.<br>
Показывает, в какую степень нужно возвести основание a, чтобы получить число b.

$$
\log_a b = c \quad \Leftrightarrow \quad a^c = b
$$

где  
- $a$ — основание ( $a>0,\ a\neq1$ ),  
- $b>0$,  
- $c$ — показатель степени.

```python
import numpy as np
```

```python
# Важно: NumPy по умолчанию использует натуральный логарифм (ln), то есть log_e(x)
x = np.array([1.0, 2.0, 3.0, 10.0])

print("x:", x)
print("np.log(x) (натуральный логарифм):", np.log(x))
print("np.log10(x) (десятичный логарифм):", np.log10(x))
```

## 1 Основные свойства логарифмов

### 1.1 Логарифм произведения
$$
\log_a(xy) = \log_a x + \log_a y
$$

**Смысл:** умножение превращается в сложение.

Пример:
$$
\log_2(8\cdot4) = \log_2 8 + \log_2 4 = 3 + 2 = 5
$$


```python
# 1) Свойство логарифма произведения: log_a(xy) = log_a(x) + log_a(y)
# Для NumPy: используем натуральный логарифм (ln), но свойство верно для любого основания.
x = np.array([2.0, 3.0, 5.0])
y = np.array([4.0, 6.0, 10.0])

left = np.log(x * y)             # ln(xy)
right = np.log(x) + np.log(y)    # ln(x) + ln(y)

print("x:", x)
print("y:", y)
print("ln(x*y):", left)
print("ln(x) + ln(y):", right)
print("Проверка (разница):", left - right)  # должно быть ~0 из-за численных погрешностей
```

### 1.2 Логарифм частного
$$
\log_a\left(\frac{x}{y}\right) = \log_a x - \log_a y
$$

**Смысл:** деление превращается в вычитание.


```python
# 2) Свойство логарифма частного: log_a(x/y) = log_a(x) - log_a(y)
x = np.array([8.0, 9.0, 100.0])
y = np.array([2.0, 3.0, 10.0])

left = np.log(x / y)           # ln(x/y)
right = np.log(x) - np.log(y)  # ln(x) - ln(y)

print("x:", x)
print("y:", y)
print("ln(x/y):", left)
print("ln(x) - ln(y):", right)
print("Проверка (разница):", left - right)  # должно быть ~0
```

### 1.3 Логарифм степени
$$
\log_a(x^k) = k \log_a x
$$

**Смысл:** показатель степени выносится вперёд.

Пример:
$$
\log_3(9^2) = 2\log_3 9 = 2\cdot2 = 4
$$


```python
# 3) Свойство логарифма степени: log_a(x^k) = k * log_a(x)
# Проверим на массиве и нескольких k.
x = np.array([2.0, 3.0, 4.0])
k = 3

left = np.log(x ** k)      # ln(x^k)
right = k * np.log(x)      # k*ln(x)

print("x:", x)
print("k:", k)
print("ln(x^k):", left)
print("k*ln(x):", right)
print("Проверка (разница):", left - right)  # должно быть ~0
```

### 1.4 Логарифм корня
$$
\log_a(\sqrt[n]{x}) = \frac{1}{n}\log_a x
$$

Потому что корень — это степень $x^{1/n}$.


```python
# 4) Свойство логарифма корня: log_a(root_n(x)) = (1/n) * log_a(x)
# root_n(x) = x^(1/n)
x = np.array([16.0, 81.0, 256.0])
n = 4

root_n = x ** (1 / n)  # четвертый корень
left = np.log(root_n)  # ln(root_n(x))
right = (1 / n) * np.log(x)

print("x:", x)
print("n:", n)
print("root_n(x):", root_n)
print("ln(root_n(x)):", left)
print("(1/n)*ln(x):", right)
print("Проверка (разница):", left - right)  # должно быть ~0
```

## 2 Специальные значения

$$
\log_a 1 = 0
$$

(потому что $a^0 = 1$)

$$
\log_a a = 1
$$

(потому что $a^1 = a$)


```python
# 2) Специальные значения: log_a(1) = 0, log_a(a) = 1
# В NumPy мы проверим через ln: ln(1)=0, ln(e)=1
print("ln(1):", np.log(1.0))
print("ln(e):", np.log(np.e))

# Для произвольного основания a проверим логарифм по формуле смены основания:
a = 5.0
print("a:", a)

log_a_1 = np.log(1.0) / np.log(a)   # log_a(1)
log_a_a = np.log(a) / np.log(a)     # log_a(a)

print("log_a(1) по формуле смены основания:", log_a_1)
print("log_a(a) по формуле смены основания:", log_a_a)
```

## 3 Формула смены основания

$$
\log_a b = \frac{\log_c b}{\log_c a}
$$

Чаще всего используют натуральный логарифм:

$$
\log_a b = \frac{\ln b}{\ln a}
$$

или десятичный:

$$
\log_a b = \frac{\log b}{\log a}
$$


```python
# 3) Формула смены основания:
# log_a(b) = ln(b) / ln(a)
a = 2.0
b = 32.0

log_a_b = np.log(b) / np.log(a)  # log_2(32)

print("a:", a, "b:", b)
print("log_a(b) = ln(b)/ln(a):", log_a_b)
print("Проверка: a^(log_a(b)):", a ** log_a_b)  # должно быть ~b
```

## 4 Важные замечания
- Логарифм определён **только для положительных чисел**.
- Основание должно быть положительным и не равным 1.
- Логарифм — монотонная функция:
  - если $a>1$ → возрастает  
  - если $0<a<1$ → убывает

```python
# 4) Важные замечания (демонстрация на практике):
# Логарифм определён только для положительных чисел.
vals = np.array([3.0, 1.0, 0.1, 0.0, -2.0])

print("vals:", vals)

# np.log для 0 и отрицательных значений даст -inf или nan и предупреждения (это ожидаемо).
with np.errstate(divide='warn', invalid='warn'):
    res = np.log(vals)

print("np.log(vals):", res)
print("Замечание: для 0 -> -inf, для отрицательных -> nan (не определено в вещественных числах).")
```

```python
# 4) Монотонность логарифма:
# Если a>1, log_a(x) возрастает по x; если 0<a<1, log_a(x) убывает по x.
x = np.array([1.0, 2.0, 4.0, 8.0])

a1 = 2.0      # a>1
a2 = 0.5      # 0<a<1

log_a1_x = np.log(x) / np.log(a1)
log_a2_x = np.log(x) / np.log(a2)

print("x:", x)
print("a1=2.0 (a>1): log_{a1}(x):", log_a1_x)
print("a2=0.5 (0<a<1): log_{a2}(x):", log_a2_x)
print("Вывод: при a>1 значения растут, при 0<a<1 — убывают.")
```
