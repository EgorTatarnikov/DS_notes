## 1. Понятие вектора

**Вектор** — это направленный отрезок, характеризующийся **длиной** (модулем) и **направлением**.

Обозначение:
$$
\vec a = \overrightarrow{AB}
$$

- **Начало** вектора — точка $A$  
- **Конец** вектора — точка $B$  
- **Длина** (модуль) вектора:
$$
|\vec a|
$$


### 1.1 Нулевой, равные и коллинеарные векторы

**Нулевой вектор** — вектор нулевой длины:
$$
\vec 0, \quad |\vec 0| = 0
$$

**Равные векторы** — векторы, у которых совпадают длина и направление.

**Коллинеарные векторы** — лежат на одной прямой или на параллельных прямых.  
Алгебраическое условие коллинеарности:
$$
\vec b = \lambda \vec a
$$
где $\lambda$ — некоторое число.


```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([2, 4, 6])

# коллинеарность: b = 2a
print("a =", a)
print("b =", b)
print("Коллинеарны ли a и b?", np.allclose(b, 2*a))

# нулевой вектор и длина
zero = np.zeros(3)
print("||a|| =", np.linalg.norm(a))
print("||0|| =", np.linalg.norm(zero))
```

## 2. Линейные операции над векторами и их свойства

В векторной алгебре основными линейными операциями являются:

- сложение $\vec a + \vec b$
- вычитание $\vec a - \vec b$
- умножение на число $\lambda \vec a$

Именно они лежат в основе понятия **линейной комбинации** и **линейной зависимости**.


### 2.1 Сумма векторов

Геометрически сумма строится **по правилу треугольника** или **по правилу параллелограмма**.

Свойства сложения:

1. **Переместительное**
$$
\vec a + \vec b = \vec b + \vec a
$$

2. **Сочетательное**
$$
(\vec a + \vec b) + \vec c = \vec a + (\vec b + \vec c)
$$

3. **Нулевой элемент**
$$
\vec a + \vec 0 = \vec a
$$


```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([-1, 0, 2])

print("a + b =", a + b)
print("(a + b) + c =", (a + b) + c)
print("a + (b + c) =", a + (b + c))
```

### 2.2 Разность векторов

Разность определяется через сложение с противоположным вектором:
$$
\vec a - \vec b = \vec a + (-\vec b)
$$


```python
print("a - b =", a - b)
print("-b =", -b)
print("a + (-b) =", a + (-b))
```

### 2.3 Умножение вектора на число

$$
\lambda \vec a = (\lambda a_1, \lambda a_2, \dots, \lambda a_n)
$$

Свойства:

1. **Распределительность относительно суммы векторов**
$$
\lambda(\vec a + \vec b) = \lambda\vec a + \lambda\vec b
$$

2. **Распределительность относительно суммы чисел**
$$
(\lambda + \mu)\vec a = \lambda\vec a + \mu\vec a
$$

3. **Сочетательность**
$$
(\lambda\mu)\vec a = \lambda(\mu\vec a)
$$


```python
lam, mu = 3, -2
print("lam*a =", lam*a)
print("lam*(a+b) =", lam*(a+b))
print("lam*a + lam*b =", lam*a + lam*b)
print("(lam+mu)*a =", (lam+mu)*a)
print("lam*a + mu*a =", lam*a + mu*a)
```

### 2.4 Линейные комбинации. Линейная зависимость и независимость

**Линейная комбинация** векторов $\vec a_1,\dots,\vec a_k$:
$$
\lambda_1\vec a_1 + \lambda_2\vec a_2 + \dots + \lambda_k\vec a_k
$$

Векторы **линейно зависимы**, если существуют числа $\lambda_1,\dots,\lambda_k$, не все равные нулю, такие что:
$$
\lambda_1\vec a_1 + \dots + \lambda_k\vec a_k = \vec 0
$$

Если такое равенство возможно только при $\lambda_1=\dots=\lambda_k=0$, то векторы **линейно независимы**.


```python
# Пример: линейная зависимость
a1 = np.array([1, 2, 3])
a2 = np.array([2, 4, 6])   # a2 = 2*a1

# Найдём коэффициенты: 2*a1 - 1*a2 = 0
comb = 2*a1 - 1*a2
print("2*a1 - a2 =", comb)
```

## 3. Базис, аффинные координаты, декартова система координат

### 3.1 Понятие базиса

**Базис** — это набор линейно независимых векторов, через который можно разложить любой вектор рассматриваемого пространства.

В $\mathbb{R}^3$ стандартный декартов базис:
$$
\vec i = (1,0,0),\quad \vec j=(0,1,0),\quad \vec k=(0,0,1)
$$

Тогда любой вектор:
$$
\vec a = x\vec i + y\vec j + z\vec k
$$
и числа $(x,y,z)$ — его координаты в данном базисе.


```python
i = np.array([1,0,0])
j = np.array([0,1,0])
k = np.array([0,0,1])

x, y, z = 2, -1, 3
a = x*i + y*j + z*k
print("a =", a)
```

### 3.2 Аффинные координаты (идея)

В учебнике вводится важная мысль: если фиксировать точку $O$ (начало отсчёта) и выбрать базис,
то положение точки $M$ можно описывать вектором
$$
\overrightarrow{OM}
$$
и его координатами.

Это удобно в задачах геометрии и в ML, где мы часто рассматриваем «точки» в пространстве признаков.


### 3.3 Длина (норма) и направляющие косинусы

Для $\vec a = (x,y,z)$:
$$
|\vec a| = \sqrt{x^2+y^2+z^2}
$$

Направляющие косинусы (косинусы углов с осями):
$$
\cos\alpha=\frac{x}{\sqrt{x^2+y^2+z^2}},\quad
\cos\beta=\frac{y}{\sqrt{x^2+y^2+z^2}},\quad
\cos\gamma=\frac{z}{\sqrt{x^2+y^2+z^2}}
$$
и выполняется:
$$
\cos^2\alpha+\cos^2\beta+\cos^2\gamma=1
$$


```python
a = np.array([2.0, -1.0, 3.0])
norm_a = np.linalg.norm(a)
cos_alpha, cos_beta, cos_gamma = a / norm_a
print("||a|| =", norm_a)
print("cosα, cosβ, cosγ =", cos_alpha, cos_beta, cos_gamma)
print("Сумма квадратов =", cos_alpha**2 + cos_beta**2 + cos_gamma**2)
```

## 4. Скалярное произведение двух векторов

### 4.1 Определение через угол

$$
\vec a\cdot\vec b = |\vec a|\,|\vec b|\cos\varphi
$$

Скалярное произведение двух векторов равно произведению длин этих векторов на косинус угла между ними, или произведению длины одного вектора и длину проекции другого вектора.

### 4.2 Формула через координаты

Если $\vec a=(a_1,a_2,a_3)$ и $\vec b=(b_1,b_2,b_3)$, то:
$$
\vec a\cdot\vec b = a_1b_1 + a_2b_2 + a_3b_3
$$


```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

dot_ab = np.dot(a, b)
print("a·b =", dot_ab)

# Угол между векторами через cos
cos_phi = dot_ab / (np.linalg.norm(a) * np.linalg.norm(b))
phi = np.arccos(np.clip(cos_phi, -1, 1))
print("cos(phi) =", cos_phi)
print("phi (рад) =", phi)
```

### 4.3 Ортогональность

Векторы ортогональны (перпендикулярны) тогда и только тогда, когда:
$$
\vec a\cdot\vec b = 0
$$


```python
u = np.array([1, 0, 0])
v = np.array([0, 2, 0])
print("u·v =", np.dot(u, v))
```

## 5. Векторное произведение векторов и свойства

Векторное произведение определено для $\mathbb{R}^3$.

### 5.1 Геометрическое определение

Модуль равен площади параллелограмма:
$$
|\vec a\times\vec b| = |\vec a|\,|\vec b|\sin\varphi
$$

Направление определяется правилом правой руки.

![Right Hand Rule](Конспект_Векторная_Алгебра_img_1.png)

### 5.2 Формула через координаты

$$
\vec a\times\vec b=
\bigl(
a_2b_3-a_3b_2,\;
a_3b_1-a_1b_3,\;
a_1b_2-a_2b_1
\bigr)
$$

Эквивалентно записи через определитель:
$$
\vec a\times\vec b=
\begin{vmatrix}
\vec i & \vec j & \vec k\\
a_1 & a_2 & a_3\\
b_1 & b_2 & b_3
\end{vmatrix}
$$


```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

cross_ab = np.cross(a, b)
print("a×b =", cross_ab)

# площадь параллелограмма
area = np.linalg.norm(cross_ab)
print("Площадь параллелограмма |a×b| =", area)
```

### 5.3 Свойства векторного произведения (ключевые)

1. **Антикоммутативность**
$$
\vec a\times\vec b = - (\vec b\times\vec a)
$$

2. **Распределительность**
$$
\vec a\times(\vec b+\vec c)=\vec a\times\vec b+\vec a\times\vec c
$$

3. **Умножение на число**
$$
(\lambda\vec a)\times\vec b = \lambda(\vec a\times\vec b)
$$

4. Векторы коллинеарны $\iff$ $\vec a\times\vec b=\vec 0$


## 6. Смешанное произведение векторов и свойства

### 6.1 Определение

Смешанное произведение трёх векторов:
$$
(\vec a,\vec b,\vec c)=\vec a\cdot(\vec b\times\vec c)
$$

### 6.2 Формула через определитель

Если $\vec a=(a_1,a_2,a_3)$, $\vec b=(b_1,b_2,b_3)$, $\vec c=(c_1,c_2,c_3)$, то:
$$
(\vec a,\vec b,\vec c)=
\begin{vmatrix}
a_1 & a_2 & a_3\\
b_1 & b_2 & b_3\\
c_1 & c_2 & c_3
\end{vmatrix}
$$


```python
a = np.array([1, 2, 3])
b = np.array([4, 0, 1])
c = np.array([2, -1, 5])

mixed = np.dot(a, np.cross(b, c))
print("(a,b,c) =", mixed)
print("Объем параллелепипеда =", abs(mixed))
```

### 6.3 Геометрический смысл и критерий компланарности

Модуль смешанного произведения равен объёму параллелепипеда, построенного на векторах:
$$
V = |(\vec a,\vec b,\vec c)|
$$

Три вектора **компланарны** (лежат в одной плоскости) тогда и только тогда, когда:
$$
(\vec a,\vec b,\vec c)=0
$$


## 7. Применение векторной алгебры в Data Science и ML

В ML почти всё можно переписать на языке векторов:

- объект данных $x$ — **вектор признаков**
- модель (в простом случае) — **линейная комбинация**
- мера близости — **скалярное произведение / косинус**
- снижение размерности — **смена базиса (PCA)**
- обучение — движение по **градиенту** (он тоже вектор)


### 7.1 Вектор признаков

Объект:
$$
x=(x_1,x_2,\dots,x_n)
$$

Например: возраст, доход, число покупок.


```python
x = np.array([25, 70000, 12], dtype=float)
print("x =", x)
```

### 7.2 Линейная модель как скалярное произведение

Линейная регрессия / логистическая регрессия:
$$
\hat y = w\cdot x + b
$$


```python
w = np.array([0.5, 0.0001, 2.0])
b = 1.0

y_hat = np.dot(w, x) + b
print("Предсказание y_hat =", y_hat)
```

### 7.3 Косинусная близость (Cosine similarity)

Очень популярна в NLP и поиске (в т.ч. при сравнении эмбеддингов).

$$
\cos\theta = \frac{x\cdot y}{\|x\|\,\|y\|}
$$


```python
def cosine_similarity(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))

u = np.array([1, 1, 0], dtype=float)
v = np.array([1, 0.9, 0], dtype=float)
print("cosine(u,v) =", cosine_similarity(u, v))
```

### 7.4 Проекция (projection) и интерпретация

Проекция вектора $x$ на направление $w$:
$$
\mathrm{proj}_w(x)=\frac{x\cdot w}{w\cdot w}\,w
$$

Интуитивно: «какая часть $x$ смотрит в сторону $w$».


```python
def projection(x, w):
    return (np.dot(x, w) / np.dot(w, w)) * w

x = np.array([2.0, 1.0, 0.0])
w = np.array([1.0, 0.0, 0.0])
print("proj_w(x) =", projection(x, w))
```

### 7.5 PCA как смена базиса (идея)

PCA ищет новые оси (новый базис), вдоль которых дисперсия данных максимальна.

Вычислительно часто используют SVD или собственные векторы ковариационной матрицы.

Ниже — минимальный пример через `sklearn`.


```python
# PCA пример (если sklearn установлен)
try:
    from sklearn.decomposition import PCA

    rng = np.random.default_rng(42)
    X = rng.normal(size=(200, 3))
    X[:, 2] = 0.2*X[:, 0] + 0.1*X[:, 1] + rng.normal(scale=0.1, size=200)  # зависимость

    pca = PCA(n_components=2)
    X2 = pca.fit_transform(X)

    print("X shape:", X.shape)
    print("X2 shape:", X2.shape)
    print("Explained variance ratio:", pca.explained_variance_ratio_)
except Exception as e:
    print("sklearn не найден или возникла ошибка:", e)
    print("Если нужно, установи: pip install scikit-learn")
```

### 7.6 Градиент как вектор и шаг градиентного спуска

Градиент функции потерь $L(w)$:
$$
\nabla L(w)=\left(\frac{\partial L}{\partial w_1},\dots,\frac{\partial L}{\partial w_n}\right)
$$

Шаг градиентного спуска:
$$
w_{new}=w-\eta\nabla L(w)
$$
где $\eta$ — learning rate.


```python
w = np.array([1.0, -2.0])
grad = np.array([0.3, -0.1])
eta = 0.5
w_new = w - eta*grad
print("w_new =", w_new)
```

### 7.7 Векторное произведение в CV/3D (и почему это встречается)

В компьютерном зрении и 3D часто нужно находить нормали к плоскости:
$$
\vec n = \vec a\times\vec b
$$
Нормаль нужна для:
- освещения (рендеринг)
- ориентации поверхности
- геометрических проверок


```python
a = np.array([1.0, 0.0, 0.0])
b = np.array([0.0, 1.0, 0.0])
n = np.cross(a, b)
print("normal =", n)
```

### 7.8 Смешанное произведение и объём (проверки в 3D)

Объём параллелепипеда:
$$
V = |\vec a\cdot(\vec b\times\vec c)|
$$

Если $V=0$, то векторы компланарны — это полезно для проверок в геометрии и 3D.


```python
a = np.array([1,0,0])
b = np.array([0,1,0])
c = np.array([0,0,1])
V = abs(np.dot(a, np.cross(b, c)))
print("V =", V)
```
