## 3. Массивы ndarray

Основной объект в NumPy — это ndarray (N-dimensional array), который:
- хранит элементы одного типа (в отличие от списков Python);
- имеет фиксированный размер (в отличие от динамических списков);
- поддерживает векторизацию (операции выполняются сразу на всем массиве, без циклов).

### Создание массивов

```python
print("\n 1D массив:\n", np.array([1, 2, 3]))        # 1D массив (вектор)
print("\n 2D массив:\n", np.array([[1, 2, 3],       # 2D массив (матрица)
                     [4, 5, 6]]))
print("\n Нули:\n", np.zeros((3, 3)))           # матрица 3x3 из нулей
print("\n Единицы:\n", np.ones((2, 4)))            # матрица 2x4 из единиц
print("\n С шагом:\n", np.arange(0, 10, 2))        # числа от 0 до 8 с шагом 2
print("\n С шагом в диапазоне:\n", np.linspace(0, 1, 5))       # 5 чисел от 0 до 1
print("\n Единичная матрица:\n", np.eye(3))                  # единичная матрица 3x3
print("\n Матрица, заполненная 7:\n", np.full((2, 3), 7))         # матрица 2x3, заполненная 7

np.random.seed(42)             # не обязательно
print("\n Случайные числа:\n", np.random.rand(2, 3))       # случайные числа [0,1)
print("\n Нормально распределенные числа:\n", np.random.randn(2, 3))      # генерирует массив случайных чисел, распределенных по стандартному нормальному закону (распределение Гаусса)
                               # (mean): 0, (std): 1
print("\n Случайные целые:\n", np.random.randint(0, 10, (3, 3)))  # случайные целые 0–9

# choice используется для генерации случайной выборки из заданного одномерного массива или диапазона чисел
# часто используется для bootstrap.
a = [1, 2, 3]
print("\n choice:\n", np.random.choice(a, size=5))  
```

## 4. Основные операции

```python
# поэлементные
x = np.array([1,2,3])
y = np.array([4,5,6])

print("x + y =", x + y)
print("x - y =", x - y)
print("x * y =", x * y)
print("x / y =", x / y)
print("x ** 2 =", x ** 2)

print("np.add =", np.add(x,y))            # аналог x + y
print("np.subtract =", np.subtract(x,y))  # аналог x - y
print("np.multiply =", np.multiply(x,y))  # аналог x * y
print("np.divide =", np.divide(x,y))      # аналог x / y
print("np.power =", np.power(x,2))        # аналог x ** 2

print("np.sum =", np.sum(x))                # сумма элементов массива
print("np.cumsum =", np.cumsum(x))          # кумулятивная сумма элементов массива


# индексация и срезы
arr = np.arange(10)

print() 
print(arr[0])        # первый элемент
print(arr[-1])       # последний элемент
print(arr[2:6])      # срез
print(arr[arr > 5])  # логическая индексация


# векторизация и Broadcasting
# Broadcasting позволяет выполнять операции над массивами разной формы без циклов

q = np.array([1,2,3])
print() 
print(q + 10)  # прибавление ко всем элементам
```

## 5. Работа с формой массива

```python
z = np.arange(6)

print(" z =", z)

z_mx = z.reshape(2,3)                         # переформатирование в другую форму
print("\n reshape 2x3:\n", z_mx)

print("\n flatten:", z_mx.flatten())          # переформатирование в вектор
print("\n transpose:\n", np.transpose(z_mx))  # транспонирование
print("\n transpose:\n", z_mx.T)              # то же самое

a1 = np.array([[1,2]])
a2 = np.array([[3,4]])
print("\n vstack:\n", np.vstack((a1,a2)))     # Соединяет таблицы одну под другой (по вертикали)
print("\n hstack:\n", np.hstack((a1.T,a2.T))) # Соединяет таблицы бок о бок (по горизонтали)
```

## 6. Логические операции

```python
arr = np.array([1,5,7,9])

print("where >5:", np.where(arr > 5))  # индексы элементов, которые больше 5
print("all >0:", np.all(arr > 0))  # True, если ВСЕ элементы больше 0
print("any <0:", np.any(arr < 0))  # True, если ХОТЯ БЫ один элемент меньше 0
print("маска 3< arr <8:", arr[(arr>3) & (arr<8)])  # возвращает элементы, которые одновременно >3 и <8
print("np.clip:", np.clip(arr, 3, 7))  # ограничивает значения массива x диапазоном [min, max]
```

## 7. Векторная алгебра

В ML:
- объект = вектор признаков
- веса = вектор
- модель часто = скалярное произведениение


```python
v1 = np.array([1,2,3])
v2 = np.array([2,4,6])
v3 = np.array([1,2,3])

print("Нулевой:", np.zeros(3))
print("Коллинеарные:", np.all(v2 == 2*v1))
print("Равные:", np.array_equal(v1,v3))

print("Сумма:", v1+v2)
print("Разность:", v2-v1)
print("Умножение на число:", 3*v1)
print("Линейная комбинация:", 2*v1 + 3*v2)

dot = np.dot(v1,v2)
print("Скалярное произведение:", dot)

length = np.linalg.norm(v1)
print("Длина вектора:", length)

cos = dot/(np.linalg.norm(v1)*np.linalg.norm(v2))
print("Cos угла:", cos)

cross = np.cross(v1,v2)
print("Векторное произведение:", cross)

v4 = np.array([1,0,1])
mixed = np.dot(v1, np.cross(v2,v4))
print("Смешанное произведение:", mixed)
```

## 8. Матрицы и линейная алгебра

В ML:
- датасет — матрица X (матрица векторов признаков)
- веса — вектор w
- модель — Xw

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print("\n Квадратная:\n", A)
print("\n Единичная:\n", np.eye(2))
print("\n Диагональная:\n", np.diag([1,2,3]))
print("\n Нулевая:\n", np.zeros((2,2)))
print("\n Ранг:", np.linalg.matrix_rank(A))

print("\n Сложение:\n", A+B)
print("\n Умножение на число:\n", 2*A)
print("\n Перемножение матриц:\n", A@B)

print("\n Определитель:", np.linalg.det(A)) # Определитель матрицы (показывает вырождена ли матрица; если 0 — обратной нет)

minor = A[1:,1:]
print("\n Базисный минор:\n", minor)
print("\n det минора:", np.linalg.det(minor))

eigvals, eigvecs = np.linalg.eig(A)
print("\n Собственные векторы:\n", eigvecs) # Собственный вектор — это такой вектор, который при умножении на матрицу не меняет направление. Он может только растянуться или сжаться.
print("\n Собственные значения:", eigvals)  # А число, на которое он растягивается — это собственное значение.

print("\n Транспонирование:\n", A.T)
print("\n Обратная матрица:\n", np.linalg.inv(A))      # Обратная матрица (A^{-1}); существует только если det(A) ≠ 0

C = np.array([[1,2],[3,4],[5,6]])
 
print("\n SVD-разложение:\n", np.linalg.svd(A))           # SVD-разложение: A = U Σ V^T (используется в PCA и понижении размерности)
print("\n Псевдообратная матрица:\n", np.linalg.pinv(C))  # Псевдообратная матрица (работает даже если матрица вырожденная или прямоугольная)
```

Умножение матрицы на вектор. Тут у x размерность (3,), а не (3,1).

Для x = np.array([...]) - ориентация строка/столбец не задается. Это просто 1-мерный вектор без ориентации.

```python
A = np.array([[1,2,3],
              [4,5,6]])
print(A.shape)

x = np.array([7,8,9])
print(x.shape)

y = A @ x
print(y, y.shape)
```

NumPy не превращает буквально (3,) в (3,1). Он просто использует специальное правило для 1-мерных массивов. То есть (3,) — это вектор без ориентации.

Результат тоже вектор без ориентации (2,)

Если сделать настоящий столбец.

```python
x = np.array([[7],
              [8],
              [9]])
print(x.shape)

y = A @ x
print(y, y.shape)
```

Вывод - столбец (2, 1).

А так задается вектор-строка.

```python
x = np.array([[7,8,9]])
print(x.shape)

# или через reshape
x = np.array([7,8,9])
x_row = x.reshape(1,-1)
print(x_row.shape)
```

### Softmax

```python
def softmax(x):
    x = x - np.max(x)           # стабилизация (убираем максимум). Функция экспоненты растёт ОЧЕНЬ быстро:
                                # значения могут выйти за пределы типа float и превращаются в inf
                                # inf / inf → nan. Softmax ломается.
                                # Softmax зависит не от абсолютных значений, а от разницы между ними.
                                # Пожтому результат не меняется.


    exp_x = np.exp(x)           # экспонента
    return exp_x / np.sum(exp_x)  # нормализация

# пример
z = np.array([2.0, 1.0, 0.1])
probs = softmax(z)

print("Softmax:", probs)
print("Сумма:", np.sum(probs))
```

### Softmax для батча (2D массив)

```python
def softmax_batch(X):
    X = X - np.max(X, axis=1, keepdims=True)  # по строкам
    exp_X = np.exp(X)
    return exp_X / np.sum(exp_X, axis=1, keepdims=True)

# пример
X = np.array([[2.0, 1.0, 0.1],
              [1.0, 3.0, 0.2]])

print(softmax_batch(X))
```

### Линейная регрессия (нормальное уравнение)
$$\theta = (X^T X)^{-1} X^T y$$


```python
X = np.array([[1,1],[1,2],[1,3]])
y = np.array([1,2,3])

theta = np.linalg.inv(X.T @ X) @ X.T @ y
print("theta:", theta)
```
