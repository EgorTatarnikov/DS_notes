## 2.1 `re`

### Краткое описание библиотеки
`re` — стандартный модуль Python для работы с **регулярными выражениями**.  
Это один из самых базовых инструментов в NLP-предобработке: он не понимает смысл текста, но отлично умеет находить и заменять **формальные шаблоны**.

### Для каких задач используется
- очистка текста;
- удаление лишних символов;
- извлечение email, ссылок, чисел, дат, хэштегов;
- нормализация пробелов и повторяющихся символов;
- разбиение строки по шаблону.

### Основные возможности
- `re.sub(pattern, repl, text)` — **заменяет** все или часть совпадений по шаблону.  
  Полезно для очистки текста: убрать цифры, ссылки, лишние пробелы, спецсимволы.
- `re.findall(pattern, text)` — **возвращает все найденные совпадения**.  
  Подходит для извлечения email, чисел, тегов, дат и других повторяющихся шаблонов.
- `re.search(pattern, text)` — **ищет первое совпадение** в строке.  
  Удобно, когда важно проверить: есть ли в тексте объект нужного типа.
- `re.split(pattern, text)` — **разбивает строку** по регулярному шаблону.  
  Можно делить текст по нескольким разделителям сразу: например, по `,`, `;`, `|`, пробелам.
- `re.match(pattern, text)` — проверяет совпадение **с начала строки**.  
  Часто используется для валидации формата.
- `re.compile(pattern)` — заранее компилирует шаблон.  
  Полезно, если одно и то же регулярное выражение будет использоваться много раз.
- Флаги `re.IGNORECASE`, `re.DOTALL`, `re.MULTILINE` — меняют режим поиска.  
  Например, `IGNORECASE` отключает чувствительность к регистру.

### Преимущества
- входит в стандартную библиотеку;
- очень гибкий;
- полезен почти в каждом NLP-пайплайне;
- хорошо решает задачи **до** этапа лингвистического анализа.

### Ограничения
- не понимает смысл текста;
- сложные шаблоны тяжело читать и поддерживать;
- регулярные выражения удобны для формальных структур, но плохо подходят для семантики;
- не заменяет токенизацию, лемматизацию, NER и другие полноценные NLP-инструменты.

### Когда стоит использовать
Когда нужна быстрая предобработка текста и извлечение формальных паттернов:
ссылок, email, чисел, тегов, дат, служебных маркеров и т. д.

### Пример установки
Установка не нужна.

### Что запомнить
- `re` — это инструмент **формальной** обработки текста, а не понимания языка.
- В NLP он чаще всего стоит в самом начале пайплайна.
- Для очистки текста особенно важны `re.sub()`, `re.findall()`, `re.search()`, `re.split()`.


### Пример 1. Очистка текста с помощью `re`


````python
import re

text = "Python для NLP в 2026 году! Подробнее: https://example.com, цена 15000 руб."

clean_text = text.lower()
clean_text = re.sub(r"https?://\S+|www\.\S+", "", clean_text) # https://... или https://... или www. ... s? - ноль или один раз. \. - точка
clean_text = re.sub(r"\d+", "", clean_text)             # \d - это любая цифра. + значит один или более раз, \d+ - одна или несколько цифр
clean_text = re.sub(r"[^a-zа-яё\s]", " ", clean_text)   # убирает любой символ, который НЕ входит в указанный набор
                                                        # префикс r"..." - это raw string (сырая строка)
                                                        # Квадратные скобки [...] - это набор символов
                                                        # ^ внутри [] - это отрицание, любой кроме
                                                        # \s - пробельные символы: включает пробел, таб, перенос строки
clean_text = re.sub(r"\s+", " ", clean_text).strip()    # \s+ - один или несколько пробельных символов

print(clean_text)
````

**Комментарий:** `re` удобно использовать для быстрой предобработки: убрать ссылки, цифры, шум и лишние символы.


### Пример 2. Извлечение email и хэштегов


````python
import re

text = "Пишите на egor@example.com или admin@test.org. Теги: #nlp #python #ml"

emails = re.findall(r"[\w\.-]+@[\w\.-]+\.\w+", text)
hashtags = re.findall(r"#\w+", text)

print("Emails:", emails)
print("Hashtags:", hashtags)
````

**Краткий вывод:** `re` — обязательный базовый инструмент, но только как часть общего NLP-стека.


## 2.2 `collections.Counter`

### Краткое описание библиотеки
`Counter` — специализированный словарь-счетчик из стандартного модуля `collections`.  
Он быстро считает, сколько раз каждый элемент встречается в последовательности.

### Для каких задач используется
- подсчет частот слов;
- подсчет частот символов;
- первичный анализ корпуса;
- поиск самых частотных токенов;
- оценка распределения слов до и после очистки текста.

### Основные возможности
- `Counter(tokens)` — строит счетчик по списку токенов.  
  На выходе получаем структуру вида `{токен: частота}`.
- `counter.most_common(n)` — возвращает **n самых частых** элементов.  
  Это одна из самых полезных операций при анализе текста.
- `counter.update(other_tokens)` — добавляет данные из нового списка токенов.  
  Подходит для накопления статистики по нескольким документам.
- `counter['слово']` — дает частоту конкретного токена.  
  Удобно для быстрой проверки словаря.
- арифметика счетчиков (`+`, `-`, `&`, `|`) — позволяет объединять и сравнивать частоты.  
  Это полезно при анализе нескольких корпусов.

### Преимущества
- очень простой;
- входит в стандартную библиотеку;
- отлично подходит для первичного анализа текста;
- помогает понять, какие слова доминируют в корпусе.

### Ограничения
- не умеет сам токенизировать или лемматизировать текст;
- не учитывает порядок слов;
- без предобработки частоты могут быть «грязными» из-за регистра, пунктуации и словоформ.

### Когда стоит использовать
Когда нужно быстро получить частоты слов, символов или токенов и сделать простой exploratory analysis.

### Пример установки
Установка не нужна.

### Что запомнить
- `Counter` — это самый быстрый путь к вопросу: **что чаще всего встречается в тексте?**
- Перед подсчетом полезно нормализовать текст: привести к нижнему регистру, убрать пунктуацию, лемматизировать.
- `most_common()` — ключевой метод для практики.


````python
from collections import Counter

text = "python nlp python machine learning nlp python data"
tokens = text.split()

counts = Counter(tokens)
print(counts)
print(counts.most_common(3))
````

**Комментарий:** после токенизации `Counter` дает мгновенный базовый срез корпуса.


## 2.3 `string`

### Краткое описание библиотеки
`string` — стандартный модуль Python с готовыми наборами символов и вспомогательными константами.  
Это не NLP-библиотека в прямом смысле, но она регулярно используется на этапе предобработки текста.

### Для каких задач используется
- удаление пунктуации;
- фильтрация цифр и служебных символов;
- построение собственных правил очистки;
- проверка принадлежности символов к определенному классу.

### Основные возможности
- `string.punctuation` — строка со стандартными знаками пунктуации.  
  Обычно используется, чтобы удалить `.,!?;:` и похожие символы.
- `string.digits` — символы `0-9`.  
  Удобно, когда нужно убрать цифры или проверить наличие чисел.
- `string.ascii_lowercase` / `string.ascii_uppercase` — латинские буквы в нижнем/верхнем регистре.  
  Полезно в фильтрации и генерации правил.
- `string.whitespace` — набор пробельных символов.  
  Помогает при нормализации пробелов и переносов.
- Совместное использование со `str.translate()` — быстрый способ удалить символы из строки.  
  Это часто быстрее и аккуратнее, чем последовательные `.replace()`.

### Преимущества
- простой и полезный в предобработке;
- входит в стандартную библиотеку;
- хорошо дополняет `re`.

### Ограничения
- это не NLP-библиотека, а низкоуровневый инструмент;
- не умеет токенизировать, лемматизировать или анализировать структуру текста;
- наборы символов часто требуют адаптации под конкретный язык и домен.

### Когда стоит использовать
Когда нужна быстрая работа с пунктуацией, цифрами и наборами символов в задаче очистки текста.

### Пример установки
Установка не нужна.

### Что запомнить
- `string` удобен там, где не нужен сложный NLP, а нужны **наборы символов**.
- Чаще всего используется вместе с `re`, `.lower()`, `.split()` и `Counter`.
- Для удаления пунктуации особенно полезна связка `string.punctuation` + `str.translate()`.


````python
import string

text = "Привет, мир! NLP: это интересно, правда?"
translator = str.maketrans("", "", string.punctuation)
print(text.translate(translator)) # преобразует строку посимвольно согласно таблице соответствий
````

**Краткий вывод:** `string` удобно использовать вместе с `re` для базовой очистки текста.


## 2.4 `NLTK`

### Краткое описание библиотеки
**NLTK** — классическая учебная библиотека для NLP.  
Она хорошо подходит для понимания базовых понятий: токенизации, стоп-слов, стемминга, лемматизации, POS-tagging и простых NLP-пайплайнов.

### Для каких задач используется
- токенизация слов и предложений;
- работа со стоп-словами;
- стемминг;
- лемматизация;
- частеречная разметка;
- построение учебных classical NLP-пайплайнов.

### Основные возможности
- `word_tokenize(text)` — разбивает текст на **слова и знаки пунктуации**.  
  Это базовая токенизация для дальнейшей обработки.
- `sent_tokenize(text)` — разбивает текст на **предложения**.  
  Полезно, если анализ ведется не по словам, а по предложениям.
- `stopwords.words('english')` — возвращает список **стоп-слов**.  
  Нужен для фильтрации частых, но малоинформативных слов.
- `PorterStemmer` / `SnowballStemmer` — делают **стемминг**, то есть грубо отрезают окончания.  
  Это быстрее лемматизации, но менее аккуратно.
- `WordNetLemmatizer` — выполняет **лемматизацию**, то есть приводит слово к словарной форме.  
  Обычно дает более интерпретируемый результат, чем стемминг.
- `pos_tag(tokens)` — назначает словам **части речи**.  
  Это полезно для базового грамматического анализа.
- Корпуса и словари NLTK — дают готовые данные для учебных задач.

### Преимущества
- отлична для обучения;
- много готовых ресурсов;
- хорошо показывает основы classical NLP;
- позволяет быстро собрать понятный учебный пайплайн.

### Ограничения
- часть ресурсов нужно скачивать отдельно;
- API местами ощущается более академическим, чем industrial;
- для production и высокопроизводительных задач часто выбирают другие инструменты;
- для современных transformer-задач NLTK уже не центральный инструмент.

### Когда стоит использовать
Когда изучаешь NLP, хочешь понять базовые методы или быстро собрать простой классический пайплайн.

### Пример установки


### Что запомнить
- `NLTK` — отличный способ **понять основы NLP руками**.
- В обучении NLTK очень полезен, но в production часто уступает `spaCy` и transformer-экосистеме.
- Разница между стеммингом и лемматизацией особенно хорошо видна именно на примерах из NLTK.

### Пример 1. Токенизация текста через `NLTK`


````python
from nltk.tokenize import word_tokenize, sent_tokenize

text = "Python хорошо подходит для NLP. NLTK удобен для обучения!"

print(sent_tokenize(text))
print(word_tokenize(text))
````

**Комментарий:** токенизация — одна из самых частых стартовых операций в NLP.


### Пример 2. Стоп-слова и стемминг


````python
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

tokens = ["this", "is", "a", "simple", "stemming", "example"]
stop_words = set(stopwords.words("english"))
filtered = [w for w in tokens if w.lower() not in stop_words]

stemmer = SnowballStemmer("english")
stems = [stemmer.stem(w) for w in filtered]

print(filtered)
print(stems)
````

### Пример 3. Лемматизация и POS-tagging


````python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

tokens = ["running", "better", "cars"]

lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(token) for token in tokens]
tags = pos_tag(tokens)

print("Леммы:", lemmas)
print("Теги:", tags)
````

**Краткий вывод:** `NLTK` отлично подходит для обучения и базовых экспериментов с classical NLP.


## 2.5 `spaCy`

### Краткое описание библиотеки
`spaCy` — современная production-friendly библиотека для industrial NLP pipeline.  
Она ориентирована не только на обучение, но и на прикладную разработку: быстрые пайплайны, токены, сущности, синтаксические связи и удобный API.

### Для каких задач используется
- токенизация;
- POS-tagging;
- лемматизация;
- named entity recognition (NER);
- dependency parsing;
- прикладной лингвистический анализ текста.

### Основные возможности
- готовые языковые пайплайны (`en_core_web_sm`, `ru_core_news_sm` и др.) — дают токенизацию, POS, лемматизацию и другие компоненты «из коробки».  
- объект `Doc` — центральная структура spaCy, представляющая разобранный текст.  
  Из него получаются токены, предложения, сущности и зависимости.
- объект `Token` — отдельный токен текста.  
  Через него доступны `token.text`, `token.lemma_`, `token.pos_`, `token.dep_`.
- объект `Span` — отрезок текста, например сущность или предложение.  
  Полезен для работы с фрагментами документа.
- `doc.ents` — список найденных именованных сущностей.  
  Используется для NER: люди, организации, даты, география и т. д.
- `token.lemma_` — лемма токена.  
  Нужна для нормализации текста.
- `token.pos_` — часть речи токена.  
  Полезна для правил и грамматического анализа.
- `token.dep_` — синтаксическая зависимость токена.  
  Помогает анализировать структуру предложения.
- `doc.sents` — итерация по предложениям.  
  Удобно для сегментации текста.

### Преимущества
- быстро;
- удобно;
- хорошо подходит для прикладной разработки;
- имеет чистую объектную модель (`Doc`, `Token`, `Span`);
- удобно комбинируется с правилами и ML-пайплайнами.

### Ограничения
- языковые модели ставятся отдельно;
- качество зависит от модели и языка;
- для отдельных языков и доменов результаты могут быть слабее, чем у специализированных transformer-моделей;
- для генерации текста spaCy не предназначен.

### Когда стоит использовать
Когда нужен быстрый и удобный NLP-пайплайн для прикладных задач: сущности, токены, POS, зависимости, предобработка документов.

### Пример установки
```bash
pip install spacy
python -m spacy download en_core_web_sm
# Для русского:
# python -m spacy download ru_core_news_sm
```

### Что запомнить
- `spaCy` — это один из лучших инструментов для **прикладного classical / industrial NLP**.
- Если нужен удобный production-friendly лингвистический пайплайн, часто начинают именно с него.
- Центральные сущности spaCy: `Doc`, `Token`, `Span`.


### Пример 1. Токенизация, лемматизация, POS


````python
import spacy

# Перед запуском:
# python -m spacy download en_core_web_sm

nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying a startup in the UK.")

for token in doc:
    print(token.text, "| lemma:", token.lemma_, "| pos:", token.pos_)
````

### Пример 2. Named Entity Recognition

NER (Named Entity Recognition) — это задача NLP, где модель находит в тексте именованные сущности и классифицирует их по типам.

Текст:
- Илон Маск основал компанию SpaceX в США в 2002 году

NER выдаст:
* Илон Маск → PERSON

````python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying a startup in the UK for $1 billion.")

for ent in doc.ents:
    print(ent.text, "|", ent.label_)
````

## 2.6 `pymorphy2, pymorphy3`

### Краткое описание библиотеки
`pymorphy2` — популярный инструмент для морфологического анализа русского языка.  
Для русскоязычного NLP он особенно важен, потому что русский язык богат словоформами, и простого `.lower()` недостаточно.

### Для каких задач используется
- морфологический разбор;
- лемматизация русского текста;
- получение грамматических признаков;
- нормализация словоформ перед частотным анализом и векторизацией.

### Основные возможности
- `MorphAnalyzer()` — создает морфологический анализатор.  
  Это основной объект библиотеки.
- `parse(word)` — возвращает возможные **разборы** слова.  
  У одного слова может быть несколько интерпретаций.
- `normal_form` — словарная форма слова.  
  Это основной инструмент русской лемматизации.
- `tag` — грамматические характеристики слова.  
  Например: часть речи, число, падеж, род.
- `score` у разбора — оценка вероятности конкретного разбора.  
  Помогает выбрать наиболее вероятную интерпретацию.
- `inflect({...})` — попытка изменить форму слова по набору грамматических признаков.  
  Это полезно не только для анализа, но и для морфологических преобразований.

### Преимущества
- очень полезен именно для русского языка;
- дает хорошую нормализацию словоформ;
- прост в использовании;
- помогает резко улучшить качество частотного анализа и поиска по словам.

### Ограничения
- не решает задачи NER, генерации и современные transformer-задачи;
- у слова может быть несколько разборов;
- выбор правильного разбора без контекста не всегда идеален;
- библиотека решает именно морфологию, а не весь NLP-стек.

### Когда стоит использовать
Когда работаешь с русским текстом и нужна хорошая лемматизация, морфологический анализ или нормализация слов перед ML.

### Пример установки
```bash
pip install pymorphy2
# При необходимости:
# pip install pymorphy2-dicts-ru
```

### Что запомнить
- Для русского языка `pymorphy2` часто полезнее простой токенизации.
- Главная практическая ценность — `normal_form` и `tag`.
- Особенно хорошо работает как этап нормализации перед `Counter`, `CountVectorizer`, `TfidfVectorizer`.
- pymorphy2 больше не поддерживается и имеет проблемы совместимости новыми верисями Python
- pymorphy3 - форк / продолжение pymorphy2 (поддерживает новые библиотеки Python)


````python
import pymorphy3

morph = pymorphy3.MorphAnalyzer()
words = ["кошки", "бежали", "лучшего", "инженеры"]

for word in words:
    p = morph.parse(word)[0]
    print(word, "->", p.normal_form, p.tag)
````

**Краткий вывод:** для русского текста `pymorphy2` — один из самых полезных инструментов.


## 3.1 `scikit-learn`

### Краткое описание библиотеки
`scikit-learn` — основная библиотека для классического машинного обучения в Python.  
В NLP она особенно важна для признакового подхода: Bag of Words, TF-IDF и baseline-классификации текста.

### Для каких задач используется
- Bag of Words;
- TF-IDF;
- классификация текста;
- baseline-модели;
- построение простых и сильных ML-пайплайнов.

### Основные возможности
- `CountVectorizer` — превращает текст в **Bag of Words**-матрицу частот.  
  Это один из базовых способов перейти от текста к числовым признакам.
- `TfidfVectorizer` — строит признаки TF-IDF.  
  Он учитывает не только частоту слова в документе, но и его редкость в корпусе.
- `Pipeline` — объединяет векторизацию и модель в единый конвейер.  
  Очень удобно для чистого и воспроизводимого кода.
- `LogisticRegression` — сильный и понятный baseline для текстовой классификации.

- `LinearSVC` — часто дает хороший результат на sparse-признаках текста.
- Метрики (`accuracy_score`, `classification_report`) — помогают оценить качество модели.

### Преимущества
- быстрый baseline;
- зрелый API;
- очень полезен для classical NLP;
- отличен для небольших и средних задач;
- помогает быстро проверить гипотезу до тяжелых DL-моделей.

### Ограничения
- не дает контекстных эмбеддингов как трансформеры;
- сильно зависит от качества предобработки и признаков;
- хуже работает там, где важен длинный контекст и смысловые связи.

### Когда стоит использовать
Когда нужна простая, интерпретируемая и сильная baseline-модель для текста: классификация, фильтрация, темы, тональность на небольших данных.

### Пример установки
```bash
pip install scikit-learn
```

### Что запомнить
- `scikit-learn` — база для **classical NLP и быстрых baseline**.
- Начинать классификацию текста часто разумно с `CountVectorizer` / `TfidfVectorizer` + простой модели.
- Даже в эпоху трансформеров этот стек остается очень полезным.


### Пример 1. Bag of Words


````python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

texts = [
    "python for nlp",
    "nlp with python",
    "machine learning and nlp"
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

bow_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print(bow_df)
````

**Комментарий:** строки — документы, столбцы — слова, значения — частоты.


### Пример 2. TF-IDF

````python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

texts = [
    "python for nlp",
    "nlp with python",
    "machine learning and nlp"
]

vectorizer = TfidfVectorizer()      # Рассчитывает TF-IDF для каждого слова в каждом документе корпуса
X = vectorizer.fit_transform(texts)

tfidf_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print(tfidf_df.round(3))
````

### Пример 3. Простая классификация текста


````python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

train_texts = [
    "I love this movie",
    "This film is great",
    "Amazing acting and story",
    "I hate this movie",
    "This film is terrible",
    "Awful and boring plot"
]
train_labels = [1, 1, 1, 0, 0, 0]

model = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", LogisticRegression())
])

model.fit(train_texts, train_labels)

for text, pred in zip(["great movie", "boring film"], model.predict(["great movie", "boring film"])):
    print(text, "->", "positive" if pred == 1 else "negative")
````

**Краткий вывод:** `scikit-learn` — обязательный инструмент для classical NLP и baseline-моделей.


## 3.2 `gensim`

### Краткое описание библиотеки
`gensim` — библиотека для распределенной семантики и тематического моделирования.  
Она особенно известна благодаря `Word2Vec`, `Doc2Vec` и `LDA`.

### Для каких задач используется
- Word2Vec;
- Doc2Vec;
- LDA;
- словари и корпуса;
- классические векторные представления текста.

### Основные возможности
- `Word2Vec` — обучает **векторы слов** по корпусу.  
  Слова, встречающиеся в похожих контекстах, получают близкие векторы.
- `Doc2Vec` — строит **векторы документов**.  
  Подходит для сравнения текстов на уровне документа.
- `Dictionary` — строит словарь токенов по корпусу.  
  Полезен при тематическом моделировании.
- `doc2bow()` — превращает документ в формат «мешка слов» для `gensim`.
- `LdaModel` — обучает LDA-модель для тематического моделирования.  
  Помогает искать скрытые темы в корпусе.
- `most_similar()` у Word2Vec — находит слова с близкими векторами.  
  Это хороший способ увидеть распределенную семантику на практике.

### Преимущества
- удобно для классических эмбеддингов;
- полезно для тематического моделирования;
- достаточно простой API;
- помогает понять эволюцию NLP до трансформеров.

### Ограничения
- это не современный transformer-стек;
- контекстные представления здесь отсутствуют;
- качество сильно зависит от корпуса и параметров обучения;
- для многих современных задач чаще выбирают sentence embeddings и трансформеры.

### Когда стоит использовать
Когда нужны Word2Vec, Doc2Vec, LDA и классические семантические подходы, особенно в учебных и аналитических сценариях.

### Пример установки
```bash
pip install gensim
```

### Что запомнить
- `gensim` — это мост между классическим NLP и современными эмбеддингами.
- `Word2Vec` хорошо показывает идею распределенной семантики.
- Для тематического моделирования `LDA` остается важным базовым инструментом.


### Пример 1. Word2Vec


````python
from gensim.models import Word2Vec

sentences = [
    ["python", "is", "great", "for", "nlp"],
    ["nlp", "uses", "text", "data"],
    ["python", "and", "gensim", "work", "well"]
]

model = Word2Vec(sentences, vector_size=20, window=2, min_count=1, workers=1)

print(model.wv["python"][:5])
print(model.wv.most_similar("nlp", topn=3))
````

### Пример 2. Doc2Vec


````python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

documents = [
    TaggedDocument(words=["python", "for", "nlp"], tags=["doc1"]),
    TaggedDocument(words=["deep", "learning", "for", "text"], tags=["doc2"]),
    TaggedDocument(words=["machine", "learning", "basics"], tags=["doc3"]),
]

model = Doc2Vec(documents, vector_size=20, window=2, min_count=1, workers=1, epochs=40)
vector = model.infer_vector(["python", "text"])
print(vector[:5])
````

### Пример 3. Тематическое моделирование LDA


````python
from gensim import corpora
from gensim.models import LdaModel

texts = [
    ["cat", "dog", "pet"],
    ["dog", "bark", "pet"],
    ["car", "engine", "wheel"],
    ["truck", "wheel", "road"]
]

dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=2, random_state=42, passes=20)

for topic_id, topic_words in lda_model.print_topics():
    print(f"Тема {topic_id}: {topic_words}")
````

**Краткий вывод:** `gensim` полезен для классических эмбеддингов и тематического моделирования.

`PyTorch` — одна из ключевых библиотек для deep learning.

### Для каких задач используется

- удобен и для исследований, и для прикладной разработки.

````python
import torch
import torch.nn as nn

embedding = nn.Embedding(num_embeddings=10, embedding_dim=4)
token_ids = torch.tensor([1, 2, 3])

embedded = embedding(token_ids)
print(embedded.shape)
print(embedded)
````

### Пример 3. Усреднение эмбеддингов предложения


````python
import torch
import torch.nn as nn

vocab = {"python": 0, "for": 1, "nlp": 2, "is": 3, "useful": 4}
sentence = ["python", "for", "nlp"]

token_ids = torch.tensor([vocab[word] for word in sentence])
embedding = nn.Embedding(num_embeddings=len(vocab), embedding_dim=8)

sentence_vectors = embedding(token_ids)
sentence_mean = sentence_vectors.mean(dim=0)

print(sentence_vectors.shape)
print(sentence_mean.shape)
print(sentence_mean)
````

````python
import tensorflow as tf
from tensorflow.keras.layers import TextVectorization

texts = tf.constant([
    "python for nlp",
    "nlp with tensorflow",
    "deep learning for text"
])

vectorizer = TextVectorization(output_mode="int")
vectorizer.adapt(texts)

print(vectorizer(texts))
````

### Пример 2. Простая нейросетевая модель


````python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import TextVectorization, Embedding, GlobalAveragePooling1D, Dense

train_texts = tf.constant(["good movie", "great film", "bad movie", "terrible film"])
train_labels = tf.constant([1, 1, 0, 0])

vectorizer = TextVectorization(output_mode="int", output_sequence_length=5)
vectorizer.adapt(train_texts)

model = Sequential([
    vectorizer,
    Embedding(input_dim=len(vectorizer.get_vocabulary()), output_dim=8),
    GlobalAveragePooling1D(),
    Dense(8, activation="relu"),
    Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.fit(train_texts, train_labels, epochs=5, verbose=0)

print(model.predict(tf.constant(["good film", "bad film"]), verbose=0))
````

**Краткий вывод:** `TensorFlow/Keras` — удобный способ быстро собрать DL-baseline для текста.

## 5.3 Контекстные эмбеддинги: BERT и современные transformer-энкодеры

## Краткое описание

Классические методы векторизации вроде **Bag of Words**, **TF-IDF** и **Word2Vec** полезны, но у них есть ограничение:  
они либо не учитывают контекст, либо учитывают его очень слабо.

**BERT** и похожие transformer-энкодеры строят **контекстные эмбеддинги**:  
одно и то же слово может получить разные векторы в зависимости от окружения.

Например:

- в предложении *"I opened an account in the bank"* слово `bank` связано с финансовой организацией;
- в предложении *"We sat on the river bank"* то же слово связано с берегом реки.

Для моделей типа BERT это уже **не один и тот же смысл**.

## Для каких задач используется

Transformer-энкодеры применяют, когда нужно работать со **смыслом текста**, а не просто с частотами слов:

- semantic search;
- поиск похожих текстов;
- кластеризация;
- reranking;
- retrieval в RAG; (sentence transformers приоритетнее, BERT - как универсальная языковая модель)
- классификация;
- извлечение признаков для downstream-задач.

## Основные возможности

- **`AutoTokenizer.from_pretrained(...)`** — загружает токенизатор выбранной модели; нужен для перевода текста в токены и ids.
- **`AutoModel.from_pretrained(...)`** — загружает encoder-модель без task-specific головы; удобно для извлечения эмбеддингов.
- **`tokenizer(...)`** — превращает строку или список строк в словарь входов модели (`input_ids`, `attention_mask` и др.).
- **`outputs.last_hidden_state`** — тензор скрытых состояний для всех токенов; основной источник токенных эмбеддингов.
- **`[CLS]`-вектор** — часто используется как представление всей последовательности, хотя не всегда является лучшим вариантом.
- **Mean pooling** — усреднение токенных векторов по маске; распространенный способ получить один embedding на предложение.
- **`torch.no_grad()`** — режим без вычисления градиентов; нужен для обычного inference.
- **`model.eval()`** — перевод модели в режим инференса; отключает поведение слоев вроде dropout.

## Преимущества

- учитывают контекст;
- дают более сильные представления текста;
- хорошо работают в современных смысловых задачах;
- применимы к предложениям, абзацам и документам.

## Ограничения

- тяжелее TF-IDF и классических методов;
- требуют больше памяти и времени;
- не всегда нужны для простых baseline-задач;
- “сырой BERT” не всегда оптимален для sentence similarity.

## Когда стоит использовать

Используй этот подход, если:

- TF-IDF уже не хватает;
- важен смысл, а не только слова;
- нужно сравнивать тексты по содержанию;
- строишь retrieval / semantic search / RAG.

## Когда можно не использовать

Необязательно сразу брать BERT, если:

- задача маленькая и учебная;
- нужен быстрый baseline;
- важна простота и интерпретируемость;
- классификация хорошо решается через `scikit-learn + TF-IDF`.

### Пример 1. Получение скрытых представлений токенов

Этот пример показывает, что encoder-модель возвращает вектор **для каждого токена**.

````python
from transformers import AutoTokenizer, AutoModel
import torch

model_name = "distilbert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
model.eval()

text = "Natural language processing is very useful in practice."

inputs = tokenizer(text, return_tensors="pt", truncation=True)

with torch.no_grad():
    outputs = model(**inputs)

token_embeddings = outputs.last_hidden_state
print(token_embeddings.shape)  # [batch_size, seq_len, hidden_size]
````

**Комментарий по результату**

Модель вернула тензор размера:

- `batch_size` — число текстов в батче;
- `seq_len` — число токенов;
- `hidden_size` — размер эмбеддинга каждого токена.

Именно из этих токенных представлений потом получают embedding слова, предложения или документа.

### Пример 2. Mean pooling для получения вектора предложения

Ниже показан один из самых частых способов получить **один вектор на предложение**.

````python
from transformers import AutoTokenizer, AutoModel
import torch

model_name = "distilbert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
model.eval()

texts = [
    "I like machine learning.",
    "I enjoy working with neural networks."
]

inputs = tokenizer(texts, return_tensors="pt", padding=True, truncation=True)

with torch.no_grad():
    outputs = model(**inputs)

token_embeddings = outputs.last_hidden_state
attention_mask = inputs["attention_mask"]

mask = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
mean_embeddings = (token_embeddings * mask).sum(dim=1) / mask.sum(dim=1)

print(mean_embeddings.shape)  # [num_texts, hidden_size]
````

**Комментарий по результату**

Здесь мы усредняем только реальные токены, а не `padding`.  
Так можно получить один embedding на текст, но для similarity-задач на практике чаще удобнее использовать `sentence-transformers`.

### Что запомнить

- BERT дает **контекстные** эмбеддинги.
- Одно и то же слово в разных предложениях может иметь разные векторы.
- `AutoTokenizer` готовит входы, `AutoModel` возвращает скрытые представления.
- `last_hidden_state` — базовый источник токенных эмбеддингов.
- Для представления предложения часто используют **mean pooling**.
- Для sentence similarity и semantic search обычно удобнее `sentence-transformers`.

## RuBERT - BERT для русского языка от SberAI
- обучен на русском (новости, Wikipedia)
- хорошо работает для:
  * NER
  * классификации
  * QA

````python
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F

# загрузка модели
tokenizer = AutoTokenizer.from_pretrained("DeepPavlov/rubert-base-cased")
model = AutoModel.from_pretrained("DeepPavlov/rubert-base-cased")

# функция для получения embedding предложения
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

    with torch.no_grad():
        outputs = model(**inputs)

    # outputs.last_hidden_state → (tokens, hidden)

    # mean pooling
    embeddings = outputs.last_hidden_state.mean(dim=1)

    # нормализация (важно для cosine similarity)
    embeddings = F.normalize(embeddings, p=2, dim=1)

    return embeddings

# тексты
text1 = "я люблю учиться"
text2 = "мне нравится ходить в школу"

# получаем embeddings
emb1 = get_embedding(text1)
emb2 = get_embedding(text2)

# cosine similarity
similarity = torch.cosine_similarity(emb1, emb2)

print(similarity.item())
````

## 5.4 Subword tokenization: почему современные модели режут текст не только на слова

## Краткое описание

Современные Transformer-модели работают не просто со словами, а с **подсловными токенами**.  
Это называется **subword tokenization**.

Идея в том, что модель должна:

- уметь работать с частыми словами как с цельными единицами;
- не ломаться на редких словах;
- справляться с новыми формами, частями слов и морфологией.

## Зачем это нужно на практике

Если бы модель использовала только цельные слова:

- словарь был бы слишком большим;
- редкие слова часто оказывались бы неизвестными;
- морфологически богатые языки обрабатывались бы хуже.

Subword tokenization решает это так:

- частые слова остаются целыми;
- редкие слова разбиваются на части;
- новые слова представимы через знакомые подчасти.

## Основные алгоритмы

### WordPiece
Часто используется в моделях семейства **BERT**.

### BPE
Часто используется в моделях семейства **GPT**, **RoBERTa**, **BART** и других.

## Основные возможности

- **`tokenizer.tokenize(text)`** — показывает, на какие токены разбивается текст.
- **`tokenizer(text)`** — сразу кодирует текст в ids и дополнительные служебные поля.
- **`convert_ids_to_tokens(...)`** — позволяет посмотреть, какие токены соответствуют числовым ids.
- **`attention_mask`** — показывает, какие токены реальные, а какие относятся к padding.
- **`truncation=True`** — обрезает слишком длинные тексты под лимиты модели.
- **`padding=True`** — выравнивает тексты в батче по длине.

## Почему это важно в реальной работе

Это критично, когда ты:

- считаешь длину текста в токенах;
- режешь документы на чанки;
- работаешь с ограничениями длины контекста;
- строишь retrieval или RAG;
- оцениваешь стоимость inference;
- не понимаешь, почему одно слово превратилось в несколько токенов.

## 5.5 Sentence-Transformers: основной инструмент для эмбеддингов предложений

## Краткое описание

`sentence-transformers` — это библиотека для получения **эмбеддингов предложений, фраз и коротких текстов**.  
На практике это один из самых удобных инструментов для задач:

- semantic similarity;
- semantic search;
- retrieval;
- clustering;
- FAQ matching;
- deduplication.

## Зачем она нужна

Можно взять обычный `transformers` и вручную построить вектор предложения из hidden states.  
Но в большинстве прикладных задач удобнее использовать именно `sentence-transformers`, потому что:

- код короче;
- модели уже подготовлены для sentence embeddings;
- меньше ручной работы;
- проще и надежнее запускать similarity-пайплайны.

## Основные возможности

- **`SentenceTransformer(model_name)`** — загружает готовую embedding-модель.
- **`.encode(texts)`** — превращает строку или список строк в эмбеддинги.
- **`.similarity(a, b)`** — вычисляет похожесть между наборами эмбеддингов.
- **Готовые модели семейства MiniLM, mpnet и multilingual** — позволяют быстро стартовать без fine-tuning.
- **Работа со списками предложений** — удобно для semantic search и массового кодирования документов.

## Преимущества

- очень удобна для sentence embeddings;
- мало кода;
- подходит для semantic search и similarity почти “из коробки”;
- хороший практический инструмент для современных NLP-задач.

## Ограничения

- это не замена всей библиотеке `transformers`;
- для сложного fine-tuning и кастомной архитектуры может понадобиться более низкий уровень;
- качество зависит от выбранной модели.

## Когда стоит использовать

Используй `sentence-transformers`, если:

- нужно сравнивать предложения по смыслу;
- нужен semantic search;
- нужен retrieval для базы знаний;
- нужно искать похожие тексты.

## Установка

````python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

sentences = [
    "Я люблю изучать обработку естественного языка.",
    "Мне нравится разбираться в NLP."
]

embeddings = model.encode(sentences)

print(embeddings.shape)
````

Каждое предложение преобразуется в один плотный вектор фиксированного размера.

**Комментарий по результату**

Более близкие по смыслу предложения получают более высокую косинусную близость.

- Это один из самых полезных инструментов для similarity и semantic search.
- Типовая схема: **текст → embedding → cosine similarity → топ похожих результатов**.
- Для большинства прикладных задач этого уже достаточно, чтобы построить сильный baseline.

## 5.6 Semantic search на эмбеддингах

## Краткое описание

**Semantic search** — это поиск не по точному совпадению слов, а по **смысловой близости**.

Вместо схемы:

> запрос → поиск по ключевым словам

используется схема:

> запрос → embedding → сравнение с embeddings документов → top-k результатов

## Где это реально используют

- поиск по базе знаний;
- поиск по документации;
- FAQ matching;
- retrieval для RAG;
- поиск похожих тикетов, описаний, резюме, вакансий.

## Почему это важно

Keyword search полезен, но у него есть ограничения:

- плохо работает с перефразировками;
- чувствителен к формулировке;
- может не уловить близкий смысл.

Semantic search:
- лучше работает со смыслом;
- находит тексты с похожей идеей;
- полезен, когда пользователь и документ используют разные формулировки.

## Основные возможности

- **предварительное кодирование документов** — эмбеддинги документов можно посчитать заранее;
- **кодирование запроса в момент поиска** — запрос превращается в embedding на лету;
- **`cosine_similarity(...)`** — базовая метрика сравнения близости;
- **top-k отбор** — выбираются самые близкие документы;
- **перевод retrieval-задачи в работу с векторами** — база для современных поисковых пайплайнов.

## Установка

````python
#!pip install sentence-transformers scikit-learn
````

### Пример. Простейший semantic search по списку текстов

````python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

documents = [
    "Python используется для анализа данных и машинного обучения.",
    "Трансформеры применяются для задач NLP и semantic search.",
    "Кошки любят лежать на диване.",
    "Scikit-learn подходит для TF-IDF и классической классификации текста."
]

query = "Какие инструменты используют для смыслового поиска в NLP?"

doc_embeddings = model.encode(documents)
query_embedding = model.encode([query])

scores = cosine_similarity(query_embedding, doc_embeddings)[0]
top_indices = np.argsort(scores)[::-1]

for i in top_indices:
    print(f"score={scores[i]:.4f} | {documents[i]}")
````

**Комментарий по результату**

Документы сортируются по близости к смыслу запроса.  
Даже если в запросе и документе разные слова, semantic search может найти правильный текст за счет близости embeddings.

### Практические замечания

- Для маленьких проектов этого уже достаточно.
- Для больших систем эмбеддинги документов обычно считают заранее.
- Для больших коллекций применяют векторные индексы и векторные БД.
- Поверх retrieval часто добавляют **reranking**.

### Что запомнить

- Semantic search — одна из самых практичных задач для современных embeddings.
- Базовая схема очень простая: **text → embedding → cosine similarity → top-k**.
- Для первого рабочего решения часто достаточно `sentence-transformers` и `scikit-learn`.
- Это один из основных мостов между NLP и современными retrieval/RAG-системами.

## 5.7 Какие эмбеддинги когда использовать

## Зачем нужен этот раздел

У новичков быстро возникает вопрос:

- когда брать **TF-IDF**;
- когда **Word2Vec**;
- когда **fastText**;
- когда **BERT**;
- когда **Sentence-BERT**.

Ниже — практическое правило выбора.

## 1. TF-IDF

Используй, если:

- нужен быстрый baseline;
- задача простая;
- важна интерпретируемость;
- нужна классическая классификация текста;
- объем данных небольшой или средний.

Подходит для:
- спама;
- простой тематики;
- фильтрации;
- baseline-классификаторов.

## 2. Word2Vec / fastText

Используй, если:

- изучаешь классические эмбеддинги;
- работаешь на уровне слов;
- хочешь понимать развитие NLP;
- нужен легкий baseline на словных представлениях.

`fastText` особенно полезен тем, что учитывает подсловную информацию и лучше переносит редкие слова.

## 3. BERT / encoder-модели из transformers

Используй, если:

- нужен контекст;
- важен смысл текста;
- решаешь современную NLP-задачу;
- строишь encoder-based pipeline.

Подходит для:
- feature extraction;
- классификации;
- retrieval-компонентов;
- NER;
- представления текста с учетом контекста.

## 4. Sentence-Transformers

Используй, если:

- нужно сравнивать предложения по смыслу;
- нужен semantic search;
- нужен retrieval;
- нужна кластеризация коротких текстов;
- нужно искать похожие документы, вопросы, описания.

## Короткое правило выбора

- **нужен быстрый baseline** → `TF-IDF`
- **нужны словные эмбеддинги** → `Word2Vec` / `fastText`
- **нужен контекстный encoder** → `transformers`
- **нужна similarity / semantic search** → `sentence-transformers`

## Что запомнить

- `TF-IDF` не устарел: это очень хороший baseline.
- `Word2Vec` и `fastText` полезны как классическая база.
- Для современных прикладных задач особенно важны `transformers` и `sentence-transformers`.
- Выбор зависит не от моды, а от типа задачи.

## 5.8 Минимальный современный стек для задач embeddings и поиска

## Вариант 1. Базовый современный стек

Подходит, если нужен минимальный набор без перегруза:

- `re`
- `scikit-learn`
- `transformers`
- `sentence-transformers`

## Вариант 2. Если нужен еще classical NLP

Подходит, если хочешь и классическую обработку текста, и современные эмбеддинги:

- `re`
- `collections`
- `NLTK` или `spaCy`
- `scikit-learn`
- `sentence-transformers`

## Вариант 3. Для русского языка

Если работаешь с русскоязычными текстами:

- `re`
- `pymorphy2`
- `scikit-learn`
- `transformers`
- `sentence-transformers`

## Практические рекомендации

### Что нужно новичку
Новичку достаточно:

- понять очистку текста;
- освоить токенизацию;
- научиться делать `Bag of Words` и `TF-IDF`;
- затем перейти к embeddings и semantic search.

### Что нужно для классического NLP
Для classical NLP обычно хватает:

- `re`
- `Counter`
- `NLTK` или `spaCy`
- `scikit-learn`

### Что нужно для современных transformer-based задач
Для современных смысловых задач особенно полезны:

- `transformers`
- `sentence-transformers`
- `cosine similarity`
- базовое понимание tokenization и embeddings

## Что запомнить

- Минимальный практический стек не обязан быть огромным.
- Для очень большого числа задач уже достаточно связки:
  **`scikit-learn + sentence-transformers + transformers`**
- Если задача простая — начинай с TF-IDF.
- Если задача про смысл, similarity, retrieval и поиск — быстро переходи к embeddings.

## 6. Практические примеры и соответствие инструментов

| Задача | Основные инструменты | Что чаще брать на практике |
|---|---|---|
| Очистка текста | `re`, `string` | `re` для шаблонов и нормализации, `string` — как вспомогательный модуль |

| Простая классификация текста | `scikit-learn`, `fastText`, `transformers` | `scikit-learn` для baseline, `transformers` — когда важен контекст |

| Retrieval / RAG-подготовка | `sentence-transformers`, `transformers`, `datasets` | эмбеддинги для поиска, `datasets` для подготовки данных |

### Что запомнить
- Для **baseline** почти всегда полезно сначала попробовать `scikit-learn` + `TF-IDF`.
- Для **контекстных представлений** и современных NLP-задач нужен `transformers`.
- Для **поиска по смыслу, similarity и retrieval** самый практичный выбор — `sentence-transformers`.
- `datasets` и tokenizers из `transformers` важны не сами по себе, а как инфраструктура современного pipeline.


## 7. Сравнение библиотек

| Библиотека | Уровень | Основные задачи | Сложность освоения | Когда применять |

| `string` | базовая | символы, пунктуация, алфавиты | низкая | вспомогательная очистка текста |
| `NLTK` | classical NLP | токенизация, стоп-слова, стемминг, POS | низкая–средняя | обучение, эксперименты, разбор базовых методов |
| `spaCy` | classical / industrial NLP | токенизация, POS, NER, dependency parsing | средняя | прикладной лингвистический pipeline |
| `pymorphy2` | русский NLP | морфология, нормальные формы, грамматические признаки | низкая–средняя | лемматизация и морфология русского текста |
| `scikit-learn` | ML | BoW, TF-IDF, классификация, baseline | средняя | быстрые baseline и классические модели |
| `gensim` | classical semantic NLP | Word2Vec, Doc2Vec, темы | средняя | классические эмбеддинги и тематическое моделирование |
| `PyTorch` | DL | тензоры, embedding-слои, свои модели | средняя–высокая | кастомный DL, fine-tuning, полный контроль |
| `TensorFlow/Keras` | DL | `TextVectorization`, простые текстовые модели | средняя | быстрые DL-прототипы |
| `transformers` | transformers | токенизаторы, encoder/decoder модели, inference, embeddings, генерация | средняя–высокая | современный NLP, classification, NER, generation, contextual embeddings |
| `datasets` | data layer | датасеты, split, map, preprocessing pipeline | средняя | подготовка данных и обучение HF-моделей |
| `sentence-transformers` | embeddings / retrieval | sentence embeddings, similarity, semantic search, retrieval | средняя | поиск по смыслу, clustering, FAQ matching, RAG |
| `fastText` | lightweight NLP | быстрые эмбеддинги и классификация, работа с subword/OOV | средняя | легкий baseline и компактные решения |

### Что особенно важно в современной практике
- `TF-IDF` и `scikit-learn` не устарели: это по-прежнему сильный baseline.
- `transformers` нужны там, где важен **контекст**.
- `sentence-transformers` особенно полезен для задач **similarity, semantic search, retrieval**.
- `fastText` все еще полезен как легковесное решение, но в смысловых задачах его часто вытесняют transformer embeddings.


## 8. Практические рекомендации

### С чего начать новичку
1. `re`, `string`, `Counter` — понять базовую очистку текста, шаблоны и простую статистику.
2. `NLTK` — разобраться, что такое токенизация, стоп-слова, стемминг, лемматизация, POS-tagging.
3. `scikit-learn` — научиться строить Bag of Words, TF-IDF и простые baseline-модели.
4. `spaCy` — перейти к более удобному прикладному NLP-пайплайну.
5. `transformers` — освоить токенизатор, готовые encoder-модели и базовый inference.
6. `sentence-transformers` — понять, как решаются задачи similarity, semantic search и retrieval.

### Какие библиотеки нужны для классического NLP
- `re`
- `string`
- `Counter`
- `NLTK`
- `spaCy`
- `pymorphy2`
- `scikit-learn`
- `gensim`

### Какие нужны для современных transformer-based задач
- `transformers`
- `datasets`
- `sentence-transformers`
- `PyTorch` или `TensorFlow/Keras`

### Минимальный базовый стек NLP-разработчика
- `re`
- `spaCy` или `NLTK`
- `scikit-learn`
- `transformers`
- `sentence-transformers`
- `pymorphy2` для русского языка

### Практическая логика выбора
- нужна **очистка** -> `re`, `string`
- нужна **простая статистика** -> `Counter`
- нужен **быстрый baseline** -> `scikit-learn`
- нужен **прикладной linguistic pipeline** -> `spaCy`
- нужен **русский морфологический анализ** -> `pymorphy2`
- нужны **классические эмбеддинги и темы** -> `gensim`
- нужны **готовые современные модели и контекстные эмбеддинги** -> `transformers`
- нужны **эмбеддинги предложений, similarity, semantic search, retrieval** -> `sentence-transformers`
- нужно **готовить HF-датасеты и preprocessing pipeline** -> `datasets`
- нужно **обучать или дообучать нейросеть** -> `PyTorch` или `TensorFlow/Keras`

### Что реально важно на практике
- Не нужно начинать сразу с самых тяжелых моделей: сильный baseline на `TF-IDF` иногда решает задачу лучше, чем ожидается.
- Для современных задач важно понимать разницу между:
  - **классическими признаками** (`BoW`, `TF-IDF`);
  - **словными эмбеддингами** (`Word2Vec`, `fastText`);
  - **контекстными эмбеддингами** (`BERT` и другие encoder-модели);
  - **эмбеддингами предложений** (`sentence-transformers`).
- Для задач поиска по смыслу и RAG особенно важны:
  - токенизаторы из `transformers`;
  - `sentence-transformers`;
  - `cosine similarity`;
  - подготовка данных через `datasets`.

### Что запомнить
- На практике хороший пайплайн строится слоями: очистка -> нормализация -> признаки/эмбеддинги -> модель -> оценка.
- Даже если финальная система будет на трансформерах, базовые инструменты предобработки все равно нужны.
- Для similarity и semantic search чаще нужен не “сырой BERT”, а именно `sentence-transformers`.


## 9. Минимальный NLP-стек: новичок / production

### 9.1 Минимальный стек для новичка
Если цель — **понять NLP и научиться решать типовые задачи**, достаточно такого стека:

- `re` — очистка и шаблоны;
- `string` — работа с пунктуацией и символами;
- `Counter` — частоты слов;
- `NLTK` — базовые методы classical NLP;
- `scikit-learn` — BoW, TF-IDF, простая классификация;
- `spaCy` — более прикладной NLP-пайплайн;
- `pymorphy2` — если работаешь с русским языком.

**Почему именно этот стек:**  
он дает хорошую учебную траекторию от простого к сложному и позволяет понять, как текст превращается в признаки и затем в модель.

### 9.2 Минимальный стек для production и прикладных задач
Если цель — **реально делать рабочие NLP-системы**, минимальный современный стек чаще выглядит так:

- `re` + `string` — очистка и нормализация;
- `spaCy` — токенизация, сущности, быстрый лингвистический пайплайн;
- `pymorphy2` — для русского языка при необходимости морфологической нормализации;
- `transformers` — токенизаторы и готовые современные модели;
- `datasets` — работа с датасетами и preprocessing pipeline;
- `sentence-transformers` — semantic search, similarity, retrieval;
- `PyTorch` — если нужен контроль над обучением и кастомными моделями.

**Почему именно этот стек:**  
он покрывает самые частые прикладные сценарии: классификация, поиск похожих текстов, извлечение информации, embeddings, inference готовых моделей и дообучение.

### 9.3 Что брать в зависимости от типа задачи

**Если задача — классическая классификация текста на небольших данных:**
- `re`
- `scikit-learn`
- иногда `pymorphy2` / `spaCy`

**Если задача — лингвистический анализ текста:**
- `spaCy`
- `NLTK`
- `pymorphy2` для русского

**Если задача — современный inference на готовой encoder/decoder-модели:**
- `transformers`

**Если задача — контекстные эмбеддинги и feature extraction:**
- `transformers`

**Если задача — semantic search / сравнение текстов по смыслу / retrieval:**
- `sentence-transformers`

**Если задача — обучение или fine-tuning модели:**
- `transformers` + `datasets` + `PyTorch`

### 9.4 Практический минимум, который действительно стоит знать
Если сократить все до ядра, то сильный базовый минимум для NLP-разработчика — это:

1. `re`
2. `Counter`
3. `spaCy`
4. `scikit-learn`
5. `transformers`
6. `sentence-transformers`
7. `pymorphy2` — если работаешь с русским текстом

### 9.5 Минимальный современный стек для embeddings и поиска
Если задача связана именно с **современными эмбеддингами, similarity и retrieval**, минимальный практически полезный стек будет таким:

- `transformers` — чтобы понимать токенизацию, encoder-модели и контекстные эмбеддинги;
- `sentence-transformers` — чтобы получать качественные sentence embeddings;
- `scikit-learn` или `torch` — чтобы считать similarity и работать с векторами;
- `datasets` — если нужно быстро готовить и обрабатывать корпус текстов.

### Что запомнить
- Новичку нужен стек, который **объясняет идеи**.
- Для production нужен стек, который **масштабируется и решает прикладные задачи**.
- На практике чаще всего используют не одну библиотеку, а их комбинацию.
- Для современных задач поиска по смыслу минимальная связка часто выглядит так:  
  `transformers` + `sentence-transformers` + similarity metric + данные.


## 10. Финальный вывод

Библиотеки Python для NLP удобно воспринимать как набор слоев:

1. **Низкоуровневая предобработка** — `re`, `string`, `Counter`
2. **Classical NLP** — `NLTK`, `spaCy`, `pymorphy2`, `scikit-learn`, `gensim`
3. **Современные эмбеддинги и inference** — `transformers`, `datasets`, `sentence-transformers`
4. **Глубокое обучение и кастомные модели** — `PyTorch`, `TensorFlow/Keras`
5. **Легковесные альтернативы** — `fastText`

На практике почти всегда используется **комбинация библиотек**:

очистка текста -> токенизация / нормализация -> признаки или эмбеддинги -> модель или similarity / retrieval -> оценка результата.

### Главная идея конспекта
Важно не просто знать названия библиотек, а понимать, **какой слой задачи** решает каждая из них:

- `re`, `string`, `Counter` — помогают подготовить текст;
- `NLTK`, `spaCy`, `pymorphy2` — помогают разбирать и нормализовать язык;
- `scikit-learn`, `gensim` — дают классические признаки, baseline-модели и словные эмбеддинги;
- `PyTorch`, `TensorFlow/Keras` — дают инфраструктуру для нейросетей и кастомного обучения;
- `transformers`, `datasets`, `sentence-transformers` — формируют современный рабочий стек для contextual embeddings, inference, semantic search и retrieval.

### Что особенно важно вынести из конспекта
- `TF-IDF` и классические baseline-подходы по-прежнему важны.
- `transformers` нужен там, где важен **контекст**.
- `sentence-transformers` особенно важен для **эмбеддингов предложений, semantic search и similarity**.
- Современный NLP почти всегда требует понимания **tokenization**, особенно subword tokenization.
- Для русского языка отдельную практическую ценность сохраняет `pymorphy2`.

### Рабочие минимумы
- **классический стек:** `re` + `spaCy`/`NLTK` + `scikit-learn`
- **современный стек:** `transformers` + `datasets` + `sentence-transformers`
- **для embeddings и поиска:** `sentence-transformers` + similarity metric
- **для русского языка:** добавить `pymorphy2`

### Финальная практическая мысль
Этот ноутбук полезно пересматривать не как список функций, а как карту инструментов:

- **что брать, когда брать и зачем брать**;
- где достаточно классического baseline;
- где уже нужен контекстный encoder;
- где лучше использовать sentence embeddings вместо ручной сборки векторов;
- как связать очистку текста, векторизацию, embeddings и прикладную задачу в один цельный NLP-пайплайн.
