# Юнит-тесты

## Что такое юнит-тесты (unit tests)?

**Юнит-тесты** — это **небольшие автоматические проверки**, которые тестируют **отдельные "единицы" кода** (от англ. *unit* — "единица"). Обычно это:
- одна функция,
- один метод класса,
- иногда — отдельный блок логики.

## Зачем они нужны?

Чтобы **гарантировать, что конкретный кусочек кода работает правильно** при разных входных данных. Если ты что-то поменяешь в проекте — юнит-тесты помогут быстро понять, **сломалось ли что-то**.

## Пример: простая функция

```python
def add(a, b):
    return a + b
```

🧪 Юнит-тест:

```python
def test_add():
    assert add(2, 3) == 5
    assert add(-1, 1) == 0
    assert add(0, 0) == 0
```

Если любая строка `assert` не выполнится — тест "упадёт", и ты сразу увидишь, где проблема.

## Инструменты для тестирования в Python

- `unittest` — встроенный модуль.
- `pytest` — простой и мощный инструмент, которым чаще пользуются в DS и backend.

## Что такое `pytest`?

`pytest` — это популярный **фреймворк для юнит-тестирования** в Python, который:
- простой в использовании,
- позволяет писать читаемые тесты без лишнего кода,
- поддерживает автопоиск файлов и функций с тестами,
- легко расширяется (фикстуры, параметры, coverage и пр.).

### Пример
```python
# test_utils.py
def test_my_function():
    assert my_function(5) == 25
```

Запуск:
```bash
pytest test_utils.py
```

## Что такое `test_utils.py`?

Это **файл с тестами**, который мы создаём. Название `test_*.py` важно — `pytest` по нему автоматически найдёт файл.

## Пример
Допустим, есть модуль `utils.py`:

```python
# utils.py
def square(x):
    return x * x
```

Мы хотим протестировать `square`. Тогда создаём рядом файл `test_utils.py`:

```python
# test_utils.py
from utils import square

def test_square_positive():
    assert square(3) == 9

def test_square_zero():
    assert square(0) == 0

def test_square_negative():
    assert square(-4) == 16
```

Каждая функция, начинающаяся с `test_`, будет считаться **отдельным тестом**.

## Как запускать?

В терминале (в той же папке, где лежат файлы):
```bash
pytest test_utils.py
```

Вывод будет примерно такой:
```
============================= test session starts =============================
collected 3 items

test_utils.py ...                                                       [100%]

============================== 3 passed in 0.01s ==============================
```

✔️ — значит, все тесты прошли.

## Полезные фишки

### 1. Проверка на ошибку (например, деление на 0)
```python
import pytest

def divide(a, b):
    return a / b

def test_divide_zero():
    with pytest.raises(ZeroDivisionError):
        divide(1, 0)
```

### 2. Параметризация
```python
import pytest

@pytest.mark.parametrize("x, expected", [(2, 4), (3, 9), (0, 0)])
def test_square(x, expected):
    assert square(x) == expected
```

### 3. Запуск всех тестов в проекте
```bash
pytest
```
Если в проекте несколько файлов вида `test_*.py`, `pytest` их всех найдёт и выполнит.
