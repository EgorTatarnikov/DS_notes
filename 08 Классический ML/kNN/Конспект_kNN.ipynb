# Алгоритм k-ближайших соседей (k-Nearest Neighbors)

## 1. Введение

Алгоритм k-ближайших соседей (kNN) — это метод обучения с учителем, который используется для задач классификации и регрессии.

Главная идея метода заключается в следующем:

> Объекты с похожими признаками имеют похожие ответы.

Алгоритм не строит параметрическую модель и не обучает веса. Он хранит обучающую выборку целиком и делает предсказания на основе расстояний до объектов обучающего множества.


## 2. Области применения

Алгоритм используется в следующих задачах:

- Классификация (например, определение спама)
- Регрессия (предсказание числового значения)
- Рекомендательные системы
- Поиск похожих объектов
- Обнаружение аномалий


## 3. Принцип работы алгоритма

Пусть имеется новый объект $x$.

Алгоритм выполняет следующие шаги:

1. Выбирается число соседей $k$.
2. Вычисляется расстояние от объекта $x$ до всех объектов обучающей выборки.
3. Выбираются $k$ объектов с наименьшими расстояниями.
4. Выполняется предсказание:
   - В задаче классификации — по большинству голосов.
   - В задаче регрессии — как среднее значение ответов соседей.

## 4. Метрики расстояния

### Евклидово расстояние

$$
d(x,y) = \sqrt{\sum_{i=1}^{d} (x_i - y_i)^2}
$$

### Манхэттенское расстояние

$$
d(x,y) = \sum_{i=1}^{d} |x_i - y_i|
$$

### Косинусная близость

$$
\cos(\theta) = \frac{x \cdot y}{\|x\| \|y\|}
$$

Часто применяется при работе с текстовыми данными.


## 5. Важные параметры алгоритма

- $k$ — количество ближайших соседей
- Метрика расстояния
- Использование весов соседей (например, обратная пропорциональность расстоянию)

Малое значение $k$ может привести к переобучению.
Большое значение $k$ может привести к чрезмерному сглаживанию.


## 6. Масштабирование признаков

Поскольку алгоритм основан на вычислении расстояний, признаки должны быть приведены к сопоставимому масштабу.

Используются:

- StandardScaler
- MinMaxScaler

Без масштабирования признаки с большим диапазоном значений будут доминировать при вычислении расстояния.


## 7. Вычислительная сложность

Обучение:
$$
O(1)
$$

Предсказание:
$$
O(n \cdot d)
$$

где:
- $n$ — число объектов обучающей выборки
- $d$ — размерность признакового пространства

Алгоритм плохо масштабируется на большие датасеты.


## 8. Преимущества

- Простота реализации
- Отсутствие этапа обучения параметров
- Универсальность
- Хорошо работает на небольших наборах данных


## 9. Недостатки

- Высокая вычислительная стоимость предсказаний
- Чувствительность к масштабу признаков
- Проблема проклятия размерности
- Необходимо хранить весь обучающий набор


## 10. Когда использовать алгоритм kNN

Алгоритм рекомендуется использовать:

- При небольших объемах данных
- В качестве базовой модели для сравнения
- При сложной форме границ между классами

Для больших датасетов следует рассматривать методы ускорения поиска соседей или использовать другие алгоритмы.


## 11. Пример использования в Python


```python
import pandas
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score 
```

```python
train_data = pandas.read_csv('Train.csv') # прочитайте данные кофейной компании для обучения и сохраните в переменную train_data
test_data = pandas.read_csv('Test.csv') # прочитайте данные кофейной компании для теста и сохраните в переменную test_data

print(train_data)
print(test_data)
```

```python
X_train = train_data.drop('Segmentation', axis=1) # подготовьте обучающую выборку и сохраните в переменную X_train
y_train = train_data['Segmentation']# сохраните целевую переменную в y_train

X_test = test_data.drop('Segmentation', axis=1) # подготовьте тестовую выборку и сохраните в переменную X_test
y_test = test_data['Segmentation']# сохраните целевую переменную в y_test

print("Размер обучающей выборки X_train:", X_train.shape) # выведите на экран размер обучающей выборки
print("Размер целевой переменной y_train:", y_train.shape) # выведите на экран размер целевой переменной на обучении
print("Размер тестовой выборки X_test:", X_test.shape) # выведите на экран размер тестовой выборки
print("Размер целевой переменной y_test:", y_test.shape) # выведите на экран размер целевой переменной на тесте
```

```python
knn = KNeighborsClassifier(n_neighbors=30) 

print(knn)
```

```python
# обучите модель по обучающей выборке классифицировать классы
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test) # запустите модель строить прогнозы на тестовой выборке

print(y_pred)
```

```python
# подсчитайте долю правильных ответов предсказанных значений целевой переменной с истинными ответами на тестовой выборке
accuracy = accuracy_score(y_test, y_pred)

print("Доля правильных ответов:", accuracy)
```
