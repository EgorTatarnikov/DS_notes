# Конспект: Метрики оценки моделей классификации и регрессии


---

## 1 Метрики для классификации

### Accuracy — Общая точность
Доля правильно предсказанных объектов:

$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$


### Precision — Прецизионность (точность положительного класса)
Из всех предсказанных положительных — сколько действительно положительных:

$$
Precision = \frac{TP}{TP + FP}
$$


### Recall (Sensitivity) — Полнота / Чувствительность / True Positive Rate
Из всех настоящих положительных — сколько модель правильно предсказала:

$$
Recall = \frac{TP}{TP + FN}
$$


### Specificity — Специфичность / True Negative Rate
Из всех настоящих отрицательных — сколько модель правильно предсказала:

$$
Specificity = \frac{TN}{TN + FP}
$$


### F1-score — Гармоническое среднее Precision и Recall
Гармоническое среднее Precision и Recall.<br>
Полезна, когда классы несбалансированы:

$$
F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}
$$


### ROC AUC (Area Under ROC Curve)
Площадь под ROC-кривой.

ROC-кривая строится по:
- True Positive Rate (Recall)
- False Positive Rate

$$
FPR = \frac{FP}{FP + TN}
$$


### PR AUC (Area Under Precision-Recall Curve)
Площадь под кривой Precision-Recall.

Используется при несбалансированных классах.

Оси:
- X: Recall
- Y: Precision


### Log Loss (Logarithmic Loss)
Наказывает модель за уверенные ошибки:

$$
LogLoss = -\frac{1}{N} \sum [y_i \log(p_i) + (1 - y_i) \log(1 - p_i)]
$$

## 2 Метрики для регрессии

### MAE (Mean Absolute Error)
Средняя абсолютная ошибка:

$$
MAE = \frac{1}{N} \sum |y_i - \hat{y}_i|
$$


### MSE (Mean Squared Error)
Средняя квадратичная ошибка:

$$
MSE = \frac{1}{N} \sum (y_i - \hat{y}_i)^2
$$


### RMSE (Root Mean Squared Error)
Корень из MSE:

$$
RMSE = \sqrt{MSE}
$$


### R² (Коэффициент детерминации)
Показывает, насколько хорошо модель объясняет вариативность (дисперсию) данных:

$$
R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
$$


### MAPE (Mean Absolute Percentage Error)
Средняя абсолютная ошибка в процентах:

$$
MAPE = \frac{1}{N} \sum \left| \frac{y_i - \hat{y}_i}{y_i} \right| \cdot 100\%
$$

## 3 Рекомендации по использованию

**Классификация**

| Сценарий                                                                       | Метрика           | Почему                                                                       |
| ------------------------------------------------------------------------------ | --------------------------------------- | ---------------------------------------------------------------------------- |
| ✅ **Классы сбалансированы**, ошибки одинаково важны                            | **Accuracy**                            | Простая и понятная, но бесполезна при дисбалансе                             |
| ⚖️ **Классы несбалансированы** (например, 95%/5%)                              | **Precision, Recall, F1-score, PR AUC** | Accuracy будет ложно высокой, F1 балансирует полноту и точность              |
| 🧪 **Важно не пропустить положительный случай** (например, рак, мошенничество) | **Recall (Sensitivity)**                | Лучше лишний раз "позвать врача", чем пропустить                             |
| 🚨 **Важно не выдать ложное срабатывание** (например, запуск тревоги)          | **Precision, Specificity**              | Ошибочные тревоги стоят дорого                                               |
| ⚖️ **Важно учитывать и FP, и FN**                                              | **F1-score**                            | Баланс между Recall и Precision                                              |
| 📊 **Нужна оценка по порогам вероятности**                                     | **ROC AUC**, **PR AUC**                 | Позволяют визуально и количественно оценить работу модели при разных порогах |
| 🧮 **Нужна оценка уверенности предсказаний**                                   | **Log Loss**                            | Чем больше уверенность в ошибочном ответе — тем выше штраф                   |


**Регрессия**

| Сценарий                                           | Метрика       | Почему                                                              |
| -------------------------------------------------- | ------------- | ------------------------------------------------------------------- |
| 🔁 **Общая ошибка**                                | **MAE**       | Простая, не чувствительна к выбросам                                |
| ⚠️ **Выбросы важны**                               | **MSE, RMSE** | Сильно штрафуют крупные ошибки, лучше при ошибках с "ценой"         |
| 📐 **Хотите сравнить с наивной моделью (средним)** | **R²**        | Показывает, насколько модель лучше "предсказания средним значением" |
| 📊 **Нужны проценты** (для менеджеров, бизнеса)    | **MAPE**      | Ошибки в % от реального значения, удобно интерпретировать           |


**Быстрые рекомендации**

| Тип задачи                          | Рекомендуемые метрики                |
| ----------------------------------- | ------------------------------------ |
| Бинарная классификация (сбаланс.)   | Accuracy, F1                         |
| Бинарная классификация (дисбаланс)  | Precision, Recall, F1, PR AUC        |
| Много-классовая классификация       | Macro F1, Accuracy, Confusion matrix |
| Регрессия (ошибка важна равномерно) | MAE                                  |
| Регрессия (большие ошибки критичны) | RMSE                                 |
| Регрессия (объяснение дисперсии)    | R²                                   |
